```python
# Drop rows missing data
df = (
    pd.read_csv("constructiondataset.csv")
    .drop(
        [
            "Index",
            "Total Existing Bldg SQFT",
            "Total Job Valuation",
            "Number Of Floors",
            "Building Valuation",
            "Electrical Valuation Remodel",
            "Unnamed: 33",
            "Contractor Full Name",
        ],
        axis=1,
    )
    .dropna(how="any")
)
```

```python
#Renaming the columns to make it consistent
df1 = df.rename(columns=lambda s: s.lower().replace(" ", "_"))
df1
```
