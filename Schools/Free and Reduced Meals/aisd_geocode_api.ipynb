```python
#create dataframe of unique school names
results_df = pd.DataFrame({"SiteName": aisd_schools_list})
results_df
```

```python
#add empty columns for appending
results_df["Lat"] = ""
results_df["Lng"] = ""
results_df["Address"] = ""
results_df.head(5)
```

```python
#find coordinates and zip code for each school

base_url = "https://maps.googleapis.com/maps/api/geocode/json"

params = params={
    "key": api_key
}

# use iterrows to iterate through pandas dataframe
for index, row in results_df.iterrows():

#get school name from dataframe 
    school_name = row["SiteName"]
    
    #add address key to to params dict
    params["address"] = school_name
    
    #assemble URL and make request
    print(f"Retrieving results for {index}: {school_name}.")
    response = requests.get(base_url, params=params).json()
          
    #extract results
    results = response["results"]
          
    try:
        results_df.loc[index, "Lat"] = results[0]["geometry"]["location"]["lat"]
        results_df.loc[index, "Lng"] = results[0]["geometry"]["location"]["lng"]
        results_df.loc[index, "Address"] = results[0]["formatted_address"]
          
    except (KeyError, IndexError):
          print("Unable to find result... skipping.")
          
print("-----------------------")
print("Data retrieval complete.")
print("-----------------------")
```

```python
#extract zipcode and store in new column of results_df
results_df["Zipcode"] = results_df['Address'].str.extract(r'(\d{5}\-?\d{0,4})')

results_df.head()
```

```python
#some number addresses were extracted before the zip code, replacing with actual zip code

results_df = results_df.replace({"Zipcode": {"10701" : "78748",
                                "12200" : "78739",
                                "12009" : "78748",
                                "11211" : "78753",
                                "10100" : "78748",
                                "12207" : "78727"}}, regex=True)
```

```python
results_df.to_csv("Resources/schools_geolocation_aisd.csv", index=False)
```
