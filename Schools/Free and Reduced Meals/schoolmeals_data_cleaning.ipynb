```python
#dependencies and set up
import os
import pandas as pd

%reload_ext nb_black
```

```python
#create paths for each csv to clean 1 at a time
path_1 = "Resources/School_Meal_Reimbursement_Year_2015-2016.csv"
path_2 = "Resources/School_Meal_Reimbursement_Year_2016-2017.csv"
path_3 = "Resources/School_Meal_Reimbursement_Year_2017-2018.csv"
path_4 = "Resources/School_Meal_Reimbursement_Year_2018-2019.csv"

#create out paths for cleaned csv files
out_path_1 = "Resources/clean_meal_reimbursement_year_2015_16.csv"
out_path_2 = "Resources/clean_meal_reimbursement_year_2016_17.csv"
out_path_3 = "Resources/clean_meal_reimbursement_year_2017_18.csv"
out_path_4 = "Resources/clean_meal_reimbursement_year_2018_19.csv"
```

#### 2015 - 2016 Data Cleaning

```python
#import and read first csv for 2015 - 2016 data
#columns were initially colapsed, use dispay.max_columns to view all
pd.set_option('display.max_columns', None)
df_1 = pd.read_csv(path_1)
df_1.head()
```

```python
#drop all columns not needed for analysis (only need quantity of students and quantity that qualifies for free or reduced lunch)
df_1.drop(columns=["ReportType", "ESC", "TDARegion", "SiteID", "SiteCounty", "BreakfastDays", "BreakfastTotal", 
                   "BreakfastADP", "BreakfastServedFree", "BreakfastServedReduced", "BreakfastServedPaid", 
                   "LunchDays", "LunchTotal", "LunchADP", "LunchServedFree", "LunchServedReduced",
                   "LunchServedPaid", "SnackDays", "SnackTotal", "SnackADP", "SnackServedFree",
                  "SnackServedReduced", "SnackServedPaid", "MilkDays", "Milktotal", "MilkADP",
                  "MilkServedFree", "MilkServedReduced", "MilkServedPaid", "BreakfastReimbursement",
                  "LunchReimbursement", "SnackReimbursement", "MilkReimbursement"], inplace=True)
```

```python
df_1.head()
```

```python
#create series of rows for schools only in Travis county
travis_co_1 = df_1.loc[df_1["CECounty"] == "TRAVIS"]

#count the number of districts to match up with ethnicity data for consistency
travis_co_1["CEName"].nunique()
```

```python
travis_co_1.dtypes
```

```python
#replace commas with "" to fix unable to parse string error
travis_co_1 = travis_co_1.replace(',','', regex=True)
travis_co_1.head()
```

```python
#change datatypes for quantity and reimbursement columns before perfoming calculations
travis_co_1[["EnrollmentQty", "FreeEligibleQty", "PaidEligibleQty", "TotalReimbursement"]] = travis_co_1[["EnrollmentQty", "FreeEligibleQty", "PaidEligibleQty", "TotalReimbursement"]].apply(pd.to_numeric)

#check datatypes
travis_co_1.dtypes
```

```python
#calcuate enrollment, free qty, reduced qty, paid qty, and total reiumbursement for each school for the entire school year
eligibility_by_school_15_16 = travis_co_1.groupby(["SiteName"]).agg({"ProgramYear": "first", 
                                      "CEID": "first", 
                                      "CEName": "first",
                                      "CECounty": "first",
                                      "TypeOfAgency": "first",
                                      "TypeOfSNPOrg": "first",
                                      "SiteName": "first",
                                      "ClaimDate": "last",
                                      "EnrollmentQty": "mean",
                                      "FreeEligibleQty": "mean",
                                      "ReducedEligibleQty": "mean",
                                      "PaidEligibleQty": "mean",
                                      "TotalReimbursement": "sum",
                                      })
eligibility_by_school_15_16
```

```python
#round all columns that were calculated
rounded_1516 = eligibility_by_school_15_16.round({"EnrollmentQty": 0,
                                   "FreeEligibleQty": 0,
                                   "ReducedEligibleQty": 0,
                                   "PaidEligibleQty": 0
                                  })

rounded_1516.head()
```

```python
#export to csv
rounded_1516.to_csv(out_path_1, index=False)
```

```python
#combine dataframes into one
AISD_15_to_19 = pd.concat([AISD_15_16, AISD_16_17, AISD_17_18, AISD_18_19])
```

```python
#reset index to remove duplicate of SiteName column
AISD_15_to_19 = AISD_15_to_19.reset_index(drop=True)
AISD_15_to_19["SiteName"].nunique()
```

```python
#checks for school name consistency across years(some schools are spelled in a slighty different way by year)
#create dataframe and output to csv to check manually
s = AISD_15_to_19.groupby(["SiteName", "ProgramYear"])["EnrollmentQty"].mean()
school_names_years = s.to_frame()

school_names_years.to_csv("Resources/school_names_years_aisd_check.csv")
```
