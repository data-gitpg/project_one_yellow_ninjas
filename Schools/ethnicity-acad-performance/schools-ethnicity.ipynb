```python
# ETHNICITY DATA

# Dependencies and setup.
import os
import csv
import pandas as pd

# Import ethnicity data.
path_eth15 = os.path.join("Enrollment Report_Statewide_Campuses_Ethnicity_2014-2015.csv")
path_eth16 = os.path.join("Enrollment Report_Statewide_Campuses_Ethnicity_2015-2016.csv")
path_eth17 = os.path.join("Enrollment Report_Statewide_Campuses_Ethnicity_2016-2017.csv")
path_eth18 = os.path.join("Enrollment Report_Statewide_Campuses_Ethnicity_2017-2018.csv")
path_eth19 = os.path.join("Enrollment Report_Statewide_Campuses_Ethnicity_2018-2019.csv")

# Convert ethnicity data to pandas data frame.
df_eth15 = pd.read_csv(path_eth15)
df_eth16 = pd.read_csv(path_eth16)
df_eth17 = pd.read_csv(path_eth17)
df_eth18 = pd.read_csv(path_eth18)
df_eth19 = pd.read_csv(path_eth19)
```

```python
# Display first dataframe to understand data structure.
df_eth15
```

```python
# Filter each dataframes to display only Travis County ethnicity data.
df_eth15_travis = df_eth15.loc[df_eth15["County Name"] == "TRAVIS COUNTY"]

# Drop Region and County columns after sorting.
df_eth15_travis = df_eth15_travis.drop(df_eth15_travis.columns[[1, 2]], axis=1)

# Create column for Grade Group for consistency. 
df_eth15_travis.insert(5, "Grade Group", "")
df_eth15_travis.head()
```

```python
# Filter each dataframes to display only Travis County ethnicity data.
df_eth16_travis = df_eth16.loc[df_eth16["County Name"] == "TRAVIS COUNTY"]

# Drop Region and County columns after sorting.
df_eth16_travis = df_eth16_travis.drop(df_eth16_travis.columns[[1, 2]], axis=1)

# Create column for Grade Group for consistency. 
df_eth16_travis.insert(5, "Grade Group", "")
df_eth16_travis.head()
```

```python
# Filter each dataframes to display only Travis County ethnicity data.
df_eth17_travis = df_eth17.loc[df_eth17["County Name"] == "TRAVIS COUNTY"]

# Drop Region and County columns after sorting.
df_eth17_clean_travis = df_eth17_travis.drop(df_eth17_travis.columns[[1, 2]], axis=1)
df_eth17_clean_travis.head()
```

```python
# Filter each dataframes to display only Travis County ethnicity data.
df_eth18_travis = df_eth18.loc[df_eth18["County Name"] == "TRAVIS COUNTY"]

# Drop Region and County columns after sorting.
df_eth18_clean_travis = df_eth18_travis.drop(df_eth18_travis.columns[[1, 2]], axis=1)
df_eth18_clean_travis.head()
```

```python
# Creating dictionary of Campus Number and Grade Group.
grade_gr_dict = df_eth18_travis.set_index(["Campus Number"])["Grade Group"].to_dict()
grade_gr_dict
```

```python
# Add Grade Group data to 2014-2015 ethnicity dataframe.
df_eth15_travis["Grade Group"] = df_eth15_travis["Campus Number"].map(grade_gr_dict)
df_eth15_clean_travis = df_eth15_travis
df_eth15_clean_travis.head()
```

```python
# Add Grade Group data to 2015-2016 ethnicity dataframe.
df_eth16_travis["Grade Group"] = df_eth16_travis["Campus Number"].map(grade_gr_dict)
df_eth16_clean_travis = df_eth16_travis
df_eth16_clean_travis.head()
```

```python
# Displaying 5th dataframe for 2018-2019 ethnicity data also reveals differences compared to other years in column order, column names and fonts.

# Filter each dataframes to display only Travis County ethnicity data.
df_eth19_travis = df_eth19.loc[df_eth19["COUNTY NAME"] == "TRAVIS COUNTY"]

# Drop Region and County columns after sorting.
df_eth19_travis = df_eth19_travis.drop(df_eth19_travis.columns[[1, 2]], axis=1)

df_eth19_travis.head()
```

```python
# Change 2018-2019 Travis County ethnicity data column name and font to match other years.
df_eth19_travis.columns = map(str.title, df_eth19_travis.columns)

df_eth19_travis.rename(columns = {"District":"District Number", "Campus":"Campus Number", "Grade Group Name":"Grade Group", "Ethnicity":"Ethnicity Name",
                          "Enrollment":"Enrollment by Ethnicity" }, inplace = True) 

df_eth19_travis.head()
```

```python
# Swap 2018-2019 Travis County ethnicity data column order (for District Number and District Name) for consistency.

cols1 = list(df_eth19_travis.columns)
a, b = cols1.index("District Number"), cols1.index("District Name")
cols1[a], cols1[b] = cols1[b], cols1[a]
df_eth19_travis = df_eth19_travis[cols1]

cols2 = list(df_eth19_travis.columns)
c, d = cols2.index("Campus Number"), cols2.index("Campus Name")
cols2[c], cols2[d] = cols2[d], cols2[c]
df_eth19_travis = df_eth19_travis[cols2]

cols3 = list(df_eth19_travis.columns)
e, f = cols3.index("Ethnicity Name"), cols3.index("Grade Group")
cols3[e], cols3[f] = cols3[f], cols3[e]
df_eth19_clean_travis = df_eth19_travis[cols3]
df_eth19_clean_travis.head()
```

```python
# Concatenate all 5 years' dataframes for Travis County ethnicity data, and reset index to avoid duplicate indices.

df_eth5yrs_travis = pd.concat([df_eth15_clean_travis, df_eth16_clean_travis, df_eth17_clean_travis, df_eth18_clean_travis, df_eth19_clean_travis])
df_eth5yrs_clean_travis = df_eth5yrs_travis.reset_index(drop=True)
df_eth5yrs_clean_travis.head()
```

```python
# Rename columns

df_eth5yrs_clean_travis.rename(columns = {"District Name":"District_Name", "District Number":"District_Number",  "Campus Name":"Campus_Name", "Campus Number":"Campus_Number", "Grade Group":"Grade_Group", "Ethnicity Name":"Ethnicity",
                          "Enrollment by Ethnicity":"Enrollment"}, inplace = True) 

df_eth5yrs_clean_travis.head()
```

```python
# Checking ethnicity types for duplicates.
df_eth5yrs_clean_travis["Ethnicity"].unique()
```

```python
# Renaming ethnicities for consistency and to avoid duplicates.

df_eth5yrs_clean_travis.loc[df_eth5yrs_clean_travis.Ethnicity.str.contains('Hispanic'),'Ethnicity'] = "Hispanic/Latino"
df_eth5yrs_clean_travis.loc[df_eth5yrs_clean_travis.Ethnicity.str.contains('Two or more races'), 'Ethnicity'] = "Multiracial"
df_eth5yrs_clean_travis.loc[df_eth5yrs_clean_travis.Ethnicity.str.contains('Two or More Races'), 'Ethnicity'] = "Multiracial"
df_eth5yrs_clean_travis.loc[df_eth5yrs_clean_travis.Ethnicity.str.contains('American Indian or Alaska Nat'),'Ethnicity'] = "American Indian or Alaska Native"
df_eth5yrs_clean_travis.loc[df_eth5yrs_clean_travis.Ethnicity.str.contains('Native Hawaiian/Other or Pacific Islander'),'Ethnicity'] = "Native Hawaiian/Other Pacific Islander"
df_eth5yrs_clean_travis.loc[df_eth5yrs_clean_travis.Ethnicity.str.contains('Native Hawaiian/Other Pacific'),'Ethnicity'] = "Native Hawaiian/Other Pacific Islander"
```

```python
# Ethnicity types - final check.
df_eth5yrs_clean_travis["Ethnicity"].unique()
```

```python
# Renaming Year with End-of-School-Year.

df_eth5yrs_clean_travis['Year'] = df_eth5yrs_clean_travis['Year'].replace('2014-2015', '2015')
df_eth5yrs_clean_travis['Year'] = df_eth5yrs_clean_travis['Year'].replace('2015-2016', '2016')
df_eth5yrs_clean_travis['Year'] = df_eth5yrs_clean_travis['Year'].replace('2016-2017', '2017')
df_eth5yrs_clean_travis['Year'] = df_eth5yrs_clean_travis['Year'].replace('2017-2018', '2018')
df_eth5yrs_clean_travis['Year'] = df_eth5yrs_clean_travis['Year'].replace('2018-2019', '2019')
df_eth5yrs_clean_travis
```

```python
# Replace MIDDLE by MS for Middle School Campus Names for consistency with other datasets.

df_eth5yrs_clean_travis['Campus_Name'] = df_eth5yrs_clean_travis['Campus_Name'].str.replace('MIDDLE', 'M S')
```

```python
# Output to a CSV file the final concatenated dataframe for 5 years of Travis County ethnicity data.

df_eth5yrs_clean_travis.to_csv("Output_ethnicity_5yrs_clean_Travis.csv", index=False)
```
