```python
import json
import requests
import pandas as pd
from config import server_key
```

```python
def extract_data(info):
    return {
        "Street Address": info["UnparsedAddress"],
        "Zip Code": info["PostalCode"],
        "List Price": info["ListPrice"],
        "Lat": info["Latitude"],
        "Long": info["Longitude"],
        "Tax Ammount": info["TaxAnnualAmount"],
        "School Dist": info["HighSchoolDistrict"],
        "List Date": info["ListingContractDate"],
    }
```

```python
url = "https://api.bridgedataoutput.com/api/v2/OData/abor_ref/Property"
payload = {
    "access_token": server_key,
    "limit": 200,
}
```

```python
# creates list to 5000 skiping by 200
count = 0
count_list = []
while count < 5000:
    count += 200
    count_list.append(count)
```

```python
# Data for 2015

data_2015 = []
house_data = []
for e in count_list:
    pull_2015 = requests.get(
        f"https://api.bridgedataoutput.com/api/v2/OData/abor_ref/Properties?access_token={server_key}&$top=200&$skip={e}&$filter=year(ListingContractDate) eq 2015"
    ).json()
    data_2015.append(pull_2015)
    data_point = pull_2015["value"]
    for e in data_point:
        house_data.append(extract_data(e))

df_2015 = pd.DataFrame(house_data).drop_duplicates()
```
